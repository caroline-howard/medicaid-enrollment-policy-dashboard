# Diagnostic Monitoring Indicators

This notebook documents the diagnostic monitoring step for the Medicaid Enrollment & Eligibility Operations Dashboard. It uses validated CMS/Data.Medicaid.gov state-month data and official Census population denominators to create descriptive monitoring outputs.

The workflow creates review flags, not problem labels. These indicators are intended to support follow-up review and policy/operations context, not causal policy evaluation.


## What did I do?

- Calculated state enrollment change from each state peak to the latest reporting month.
- Confirmed population-adjusted context metrics using official Census denominators.
- Created monitoring review flags for unusual state/month changes, latest-month preliminary status, and high missingness.
- Wrote dashboard-ready monitoring tables to `outputs/dashboard_tables/`.


```python
from pathlib import Path
import csv

project_root = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
output_dir = project_root / "outputs" / "dashboard_tables"

expected_outputs = [
    "enrollment_change_from_peak.csv",
    "state_population_adjusted_context.csv",
    "monitoring_review_flags.csv",
    "state_monitoring_summary.csv",
    "national_monitoring_summary.csv",
]

for filename in expected_outputs:
    path = output_dir / filename
    with path.open(newline="") as file:
        row_count = sum(1 for _ in csv.DictReader(file))
    print(f"{filename}: {row_count:,} rows")
```

## Why did I do it?

The dashboard needs to show more than static charts. A monitoring workflow helps users identify state/month records that may warrant closer source review, while keeping interpretation neutral and descriptive.


## What does it mean for reporting and policy monitoring?

The outputs support state comparison, enrollment trend monitoring, eligibility operations monitoring, population-adjusted context, and data-quality review. Review flags should guide questions for follow-up, not conclusions about performance or policy impact.


## Data limitations

- Public aggregate data cannot show beneficiary-level experiences.
- The data do not include claims, utilization, cost, diagnosis, or county-level detail.
- Latest-month records may be preliminary.
- Population denominators are annual state estimates and are used only for descriptive context.
- Monitoring flags are not causal findings or performance scores.
